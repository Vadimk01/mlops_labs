# Імпорти 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Налаштування відображення
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)

sns.set(style="whitegrid")
```

# Завантаження даних

```python
df = pd.read_csv("../data/raw/creditcard.csv")
df.head()
```

# Базова інформація про датасет

```python
print("Розмір датасету:", df.shape)
print("\nПерші 5 рядків:")
display(df.head())

print("\nОпис числових ознак:")
display(df.describe())
```

# Перевірка типів даних і пропусків

```python
print("Інформація про датасет:")
df.info()
```

```python
print("Кількість пропусків у кожному стовпці:")
missing_values = df.isnull().sum()
display(missing_values[missing_values > 0])

print("\nЗагальна кількість пропущених значень:", df.isnull().sum().sum())
```

Щоб побачити всі стовпці, навіть якщо пропусків немає:

```python
missing_table = pd.DataFrame({
    "column": df.columns,
    "missing_count": df.isnull().sum().values,
    "missing_percent": (df.isnull().sum().values / len(df)) * 100
})

display(missing_table.sort_values(by="missing_count", ascending=False))
```

# Перевірка цільової змінної

У цьому датасеті цільова змінна — Class:

- 0 — звичайна транзакція

- 1 — шахрайська транзакція

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="Class", data=df)
plt.title("Target Distribution (Class)")
plt.xlabel("Class (0 = Normal, 1 = Fraud)")
plt.ylabel("Count")
plt.show()
```

# Візуалізація Target Distribution

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="Class", data=df)
plt.title("Target Distribution (Class)")
plt.xlabel("Class (0 = Normal, 1 = Fraud)")
plt.ylabel("Count")
plt.show()
```

Bідсотковий розподіл: 

```python
class_percent = df["Class"].value_counts(normalize=True) * 100

plt.figure(figsize=(6, 4))
class_percent.sort_index().plot(kind="bar")
plt.title("Target Distribution in Percentage")
plt.xlabel("Class")
plt.ylabel("Percentage")
plt.xticks(rotation=0)
plt.show()

print(class_percent)
```

# Матриця кореляції

```python
plt.figure(figsize=(18, 12))
corr_matrix = df.corr()

sns.heatmap(
    corr_matrix,
    cmap="coolwarm",
    center=0,
    annot=True,
    fmt=".2f"
)

plt.title("Correlation Matrix")
plt.show()
```

Кореляція ознак із цільовою змінною:

```python
corr_with_target = df.corr(numeric_only=True)["Class"].sort_values(ascending=False)

plt.figure(figsize=(8, 10))
corr_with_target.drop("Class").plot(kind="barh")
plt.title("Feature Correlation with Target (Class)")
plt.xlabel("Correlation")
plt.ylabel("Features")
plt.show()

display(corr_with_target)
```
